# Поиск наборов данных
1. Найти 4 набора данных в открытых источниках, например, на сайте kaggle.com.

    Требования к наборам данных
    
    * данные табличные
    * 2 из 4-х наборов данных можно использовать для решения задачи регрессии оставшиеся 2 набора - для решения задачи классификации

2. Описать каждый из наборов данных по следующим пунктам

    * ссылка на набор данных в интернете
    * размер данных: количество объектов, количество признаков
    * для решения какой задачи может быть использован набор, какой признак является целевым, для задач классификации возможные классы для целевого признака
    * перечислить нецелевые столбцы и определить их типы

## Регрессия
### House Prices
> https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data
* Кол-во признаков: 81
* Кол-во объектов: 1460
* Задачи: спрогнозировать цену продажи домов
* Целевой признак: SalePrice
* Возможные классы целевого признака: -
* Нецелевые столбцы:
    * MSSubClass: класс здания. *Номинальные* 
    * MSZoning: Общая классификация зонирования. *Номинальные* 
    * LotFrontage: погонные футы улицы, соединенной с участком. *Количественные* 
    * LotArea: размер участка в квадратных футах. *Количественные* 
Улица: Тип подъезда к дороге *Номинальные* <br>
    * Переулок: Тип доступа к переулку *Номинальные* 
    * LotShape: Общий вид объекта недвижимости *Номинальные* 
    * LandContour: Ровность участка *Номинальные* 
    * LandSlope: Уклон участка *Количественные* 
    * Район: физические местоположения в пределах города Эймс. *Количественные* 

### Student Performance
> https://www.kaggle.com/datasets/haseebindata/student-performance-predictions/data
* Кол-во признаков: 9
* Кол-во объектов: 10
* Задачи: оценивать и прогнозировать результаты учащихся на основе различных факторов, которые могут повлиять на академические успехи
* Целевой признак: FinalGrade
* Возможные классы целевого признака: -
* Нецелевые столбцы:
    * StudentID: уникальный идентификатор для каждого студента. *Количественные*
    * Имя: Имя студента. *Номинальные*
    * Пол: Пол учащегося (мужчина/женщина). *Бинарные*
    * AttendanceRate: процент занятий, которые учащийся посетил. *Количественные*
    * StudyHoursPerWeek: количество часов, которые студент тратит на обучение каждую неделю. *Количественные*
    * Предыдущая оценка: оценка, которую студент получил в предыдущем семестре (из 100). *Порядковые* 
    * Внеклассная деятельность: количество внеклассных мероприятий, в которых участвует учащийся. *Порядковые*
    * Родительская поддержка: Качественная оценка уровня поддержки, оказываемой родителями учащегося (Высокий/Средний/Низкий). *Номинальные*
    * FinalGrade: итоговая оценка учащегося (из 100), которая служит целевой переменной для прогнозирования. *Порядковые*

## Классификация
### Iris Species
> https://www.kaggle.com/datasets/uciml/iris
* Кол-во признаков: 6
* Кол-во объектов: 150
* Задачи: использование множественных измерений в таксономических задачах, определение вида ириса
* Целевой признак: Species
* Возможные классы целевого признака: Iris-setosa, Iris-versicolor, Iris-virginica
* Нецелевые столбцы:
    * Id: уникальный идентификатор. *Количественные*
    * SepalLengthCm: Длина чашелистика см. *Количественные*
    * SepalWidthCm: Ширина чашелистика см. *Количественные*
    * PetalLengthCm: Длина лепестка см. *Количественные*
    * PetalWidthCm: Ширина лепестка см. *Количественные*

### Youtube Comments Spam
> https://www.kaggle.com/datasets/ahsenwaheed/youtube-comments-spam-dataset
* Кол-во признаков: 6
* Кол-во объектов: 1956
* Задачи: определение спамовых сообщений
* Целевой признак: class
* Возможные классы целевого признака: 1 указывает на спам-комментарий, а 0 – на законный комментарий.
* Нецелевые столбцы:
    * comment_id: уникальный идентификатор. *Количественные*
    * Author: автор. *Количественные*
    * Date: дата. *Количественные*
    * Content: содержание коммента. *Количественные*
    * video_name: название видео. *Количественные*
